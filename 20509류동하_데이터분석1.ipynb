### enumerate()함수: 리스트의 요소를 반복할 때 현재 인덱스가 몇번째인지 확인해야하는 경우에 사용하는 함수

```python
ex_list = ['요소1', '요소2', '요소3']

print(ex_list)
print()

#enumerate()함수 사용해 출력
print(enumerate(ex_list))
print()

#list()함수로 변경후 출력
print(list(enumerate(ex_list)))
print()

#for문과 조합
for i, value in enumerate(ex_list):
  print("{}번째 요소는 {}".format(i, value))
```

```python
olympic =[
          [1,'미국',46,37,38,121],
          [2,'영국',27,23,17,67],
          [3,'중국',26,18,26,70],
          [4,'러시아',19,18,19,56],
          [5,'독일',17,10,15,42],
          [6,'일본',12,8,21,41],
          [7,'프랑스',10,18,14,42],
          [8,'대한민국',9,3,9,21]
        ]
```

```python
print(olympic[0])
```

```python
print(olympic[0][1])
```

```python
nation=[]
for i in olympic:
  nation.append(i[1])

print(nation)
```

```python
gm=[]
for i in olympic:
  gm.append(i[2])

print(gm)
```

```python
sm=[]
for i in olympic:
  sm.append(i[3])

print(sm)
```

```python
bm=[]
for i in olympic:
  bm.append(i[4])

print(bm)
```

```python
import matplotlib.pyplot as plt

plt.title('나라별 금메달 개수')
plt.plot(nation, gm, color='gold', label='gold')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('나라별 은메달 개수')
plt.plot(nation, sm, color='silver', label='silver')
plt.legend()
plt.show()
```
