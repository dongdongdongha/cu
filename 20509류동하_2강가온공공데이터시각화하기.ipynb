```python
x = [1,2,3,4,5]
y = [2,4,8,16,32]
z = [3,9,27,81,243]

plt.title(" ")
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x,y,label = '2 ')
plt.plot(x,z,label = '3 ')
plt.legend(loc=2)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [n*2 for n in x]

plt.plot(x,y,color = 'red', marker = 'o', markersize = '10',linestyle = ":")

plt.grid()
plt.axis([0,6,0,12])

plt.show()
```

```python
import csv
```

```python
#   ,   

f = open('weather.csv', encoding='cp949')
data=csv.reader(f)
header=next(data)
data=list(data)
f.close()
```

```python
data[:2]
```

```python
header
```

```python
h = []
l = []

for r in data:
  if r[2] != '' and '08-15' in r[0]:
    h.append(float(r[2]))
    l.append(float(r[3]))

plt.figure(figsize = ([15,2]))
plt.title(" / ")
plt.plot(h,linestyle = '-')
plt.plot(l,linestyle = '-')
plt.show()

```
