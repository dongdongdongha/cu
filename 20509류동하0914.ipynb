```python
result = []
for row in data:
  if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '09':
    if row[1] != '':
      result.append(float(row[1]))
print(result)
```

```python
sum=0
for i in result:
  sum += i
print(' ', len(result), " ,  ", sum/len(result),'')
```

```python
import matplotlib.pyplot as plt

plt.title('   ')
plt.plot(result, color='orange', label='')
plt.legend()
plt.show()
```
