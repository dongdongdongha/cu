```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

```python
import csv
f = open('시계열통계.csv',encoding='utf-8')
data = csv.reader(f)
header = next(data)
data = list(data)
f.close()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 공항과 월 데이터 추출
airports = [row[0] for row in data[1:]]
months = data[0][2:-1]

# 공항별 월별 국내선 이용수 추출
domestic_airports = {}

for row in data[1:]:
    airport_name = row[0]
    domestic_flights = [int(count.replace(',', '')) for count in row[2:-1]]
    domestic_airports[airport_name] = domestic_flights

# 막대그래프 시각화
plt.figure(figsize=(12, 8))
bar_width = 0.2
bar_positions = np.arange(len(months))

for airport, counts in domestic_airports.items():
    plt.bar(bar_positions, counts, width=bar_width, label=airport)

plt.xlabel('월')
plt.ylabel('이용수')
plt.title('월별 국내 공항 이용수')
plt.xticks(bar_positions, months)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np


# 월별 국제선 이용수 추출
international_flights = {row[0]: [int(count.replace(',', '')) for count in row[2:-1]] for row in data[1:]}
plt.figure(figsize=(12, 8))
# 막대그래프 시각화
bar_positions = np.arange(len(months))

for airport, counts in international_flights.items():
    plt.bar(bar_positions, counts, label=f'{airport} (국제)')

plt.xlabel('월')
plt.ylabel('이용수')
plt.title('월별 국제 공항 이용수')
plt.xticks(bar_positions, months)
plt.legend()
plt.show()
```
