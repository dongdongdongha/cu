```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'NanumBarunGothic')
```

```python
import csv
f = open('weather.csv', encoding = 'euc-kr')
data = csv.reader(f)
header = next(data)
data = list(data)
f.close()
```

```python
print(header)
for r in data:
  print(r)
```

```python
mt = -99
md = ''

for r in data:
  if r[-2] != '':
    if mt < float(r[-2]):
      mt = float(r[-2])
      md = r[0]
print('기상관측 이래 서울의 최고기온은', md,'날',mt,'도이다.')
```

```python
h=[]
l=[]

for r in data:
  if r[2] != '' and r[3] != '':
    if r[0].split('-')[1]=='08' and r[0].split('-')[2]=='15':
      h.append(float(r[2]))
      l.append(float(r[3]))

plt.figure(figsize = (7,4))
plt.title('8월 15의 최고기온과 최저기온의 그래프')
plt.plot(h,'r',label='최고기온')
plt.plot(l,'b',label='최저기온')
plt.legend()
plt.show()
```

```python
month = [[] for i in range(12)]
for r in data:
  if r[2]!='':
    month[int(r[0].split('-')[1])-1].append(float(r[2]))

plt.boxplot(month, showfliers = False)
plt.show()
```

```python
import csv
f = open('train.csv', encoding = 'euc-kr')
data = csv.reader(f)
header = next(data)
data = list(data)
f.close()
```

```python
a = {'생존남성':0,'사망남성':0,'생존여성':0,'사망여성':0}

for r in data:
  if r[4] == 'male':
    if r[1] == '0':
      a['사망남성'] += 1
    else:
      a['생존남성'] += 1
  else:
    if r[1] == '0':
      a['사망여성'] += 1
    else:
      a['생존여성'] += 1

print(a)

plt.figure(figsize = (7,4))
plt.legend()
plt.show()
```
